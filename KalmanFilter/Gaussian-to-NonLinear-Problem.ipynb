```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
```

```python
# Generate 10000 random values drawn from a Gaussian distribution
x = np.random.normal(loc=0.0, scale=3.0, size=10000)
sns.distplot(x);
```

```python
# Map Gaussian distribution to a non-linear function: y = arctan(x)
y = np.arctan(x)
sns.distplot(y)
```

```python
# Lineplot
ax = sns.lineplot(x=x, y=y)
```
